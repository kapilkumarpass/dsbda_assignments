```python
import pandas as pd
ds = pd.read_csv("academic_performance.csv")
```

```python
ds
```

```python
for i in list(ds.columns):
    print(i)
    boolean = pd.isnull(ds[i])
    if len(ds[boolean])==0:
        print("No empty values")
        print(ds[i].value_counts())
    else:
        print(ds[boolean])
    print('----------------------------------------------------')
```

```python
def grade_convert(x):
    if x=="F":
        return 1
    else:
        return 2
        

ds['Semester']=ds['Semester'].apply(grade_convert)
```

```python

def satisfaction(x):
    if x=='Good':
        return 1
    else:
        return -1
    
ds['ParentschoolSatisfaction']=ds['ParentschoolSatisfaction'].apply(satisfaction)
```

```python
absent_days = list(ds['StudentAbsenceDays'].unique())
def absent(x):
    return absent_days.index(x)
ds['StudentAbsenceDays']=ds['StudentAbsenceDays'].apply(absent)
```

```python

def survey(x):
    if x=='Yes':
        return 1
    else:
        return 0
    
ds['ParentAnsweringSurvey']=ds['ParentAnsweringSurvey'].apply(survey)
```

```python
int(ds['Discussion'].mean())
```

```python
dis_avg=int(ds['Discussion'].mean())
def discussion(x):
    if x<15:
        return dis_avg
    else:
        return x
    
ds['Discussion']=ds['Discussion'].apply(discussion)
```

```python
int(ds['VisITedResources'].mean())
```

```python
vis_avg=int(ds['VisITedResources'].mean())
def visited_resource(x):
    if x<20:
        return vis_avg
    else:
        return x
    
ds['VisITedResources']=ds['VisITedResources'].apply(visited_resource)
```

```python
int(ds['AnnouncementsView'].mean())
```

```python
ann_avg=int(ds['AnnouncementsView'].mean())
def ann_view(x):
    if x<5:
        return ann_avg
    else:
        return x
    
ds['AnnouncementsView']=ds['AnnouncementsView'].apply(ann_view)
```

```python
def grade_convert(x):
    y=0
    x=x[::-1]
    for i in range(len(x)):
        if x[i]!='-':
            y+=(int(x[i])*(10**i))
        else:
            break
    return y
        

ds['GradeID']=ds['GradeID'].apply(grade_convert)
```
